```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
price=[0.5,0.6,0.7,0.8,1.1,1.2,1.3,1.5,2,2.5,3,4]
data={}
price_row=pd.read_excel('Total.xlsx',skiprows=1,parse_cols=[1])
for i,p in enumerate(price):
    data[p]=pd.read_excel('Total.xlsx',skiprows=1,parse_cols=[2+3*i,3+3*i,4+3*i])
```

```python
for p in price:
    data[p]['PT']=((data[p]['01 Prob. PT']>data[p]['02 Prob. PM']) * (data[p]['01 Prob. PT']>data[p]['03 Prob. SM']))
```

```python
sales={}
number={}
for p in price:
    sales[p]=sum(price_row['cost_PT']*data[p]['PT']*p)
    number[p]=sum(data[p]['PT'])
```

```python
fig, axe = plt.subplots()
axe.plot(price,sales.values())
axe.set_ylabel('Sales', color='b')
axe.set_xlabel('Price variation')
ax2 = axe.twinx()
ax2.plot(price,number.values(),'r')
ax2.set_ylabel('Costumers', color='r')
axe.grid()
plt.savefig('sales.eps', format='eps', dpi=1000)
plt.show()
```
